**웅이**

```python
positive = 192
negative = 108
total = int(positive) + int(negative)

positive = 100*float(positive)/float(total) 
negative = 100*float(negative)/float(total) 

import matplotlib.pyplot as plt

ratio = [positive, negative]
labels = ['positive', 'negative']
colors = ['#c1abff', '#d8e1c5']
plt.pie(ratio, labels=labels, autopct='%.1f%%',shadow=True,colors=colors, startangle=90)
plt.tight_layout()
plt.axis("equal")
# plt.title("woong-e")
plt.show()

```
